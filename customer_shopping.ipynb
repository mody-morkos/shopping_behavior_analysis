## Calculate value counts for string columns

```python
l = list(df.columns)
for col in l:
    if str(df[col].dtypes) =='object':
        print(df[col].value_counts(),f"\nNumber of {col} : ",df[col].nunique(),'\n')
```

## Perform initial data inspection (structure and missing values)

```python
df['Review Rating'].isna().sum()
```

## Handle review ratings for each category

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

## Convert "Frequency of Purchases" text values to numeric using mapping

```python
df.head(10)
```

```python
df['Frequency of Purchases'].unique().tolist()
```

```python
Frequency_map = {
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90,
    'Fortnightly':14
}
```

```python
df['Frequency day']=df['Frequency of Purchases'].map(Frequency_map)
```

```python
df['Frequency day']
```

## Clean and standardize column names formatting

```python
df.columns= df.columns.str.replace(' ','_').str.lower()
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
df.to_excel('C:/Users/Dataشركه/Desktop/customer.xlsx' , index=False)
```

## What is the total revenue generated by male vs. female customers?


```python
df.groupby('gender')[['purchase_amount']].sum()
```

## Which customers used a discount but still spent more than the average purchase amount? 


```python
df[(df['purchase_amount'] > df['purchase_amount'].mean() ) & (df['discount_applied']=='Yes')][['customer_id','purchase_amount']]
```

## another way

```python
df.query("discount_applied == 'Yes' and purchase_amount > purchase_amount.mean()")[['customer_id','purchase_amount']].head()
```

## For customers with multiple orders, sum total purchase amounts and select those exceeding the overall mean who used discounts

```python
g1=df[(df['discount_applied']=='Yes')].groupby('customer_id').agg(purchase_amount_sum = ('purchase_amount','sum')).reset_index()
m= df['purchase_amount'].mean()
g1[g1['purchase_amount_sum']>m]
```

## Which are the top 5 products with the highest average review rating?


```python
df.groupby('item_purchased')['review_rating'].mean().round(2).sort_values(ascending=False).head(5).reset_index()
```

## Compare the average Purchase Amounts between Standard and Express Shipping.

```python
shipping_type = ['Standard','Express']
df[df['shipping_type'].isin(shipping_type)].groupby('shipping_type')['purchase_amount'].mean().round(2).reset_index()
```

## Do subscribed customers spend more? Compare average spend and total revenue between subscribers and non-subscribers

```python
df.groupby('subscription_status').agg({'customer_id':'count',
                                      'purchase_amount':['sum','mean']}).round(2)
```

## Which 5 products have the highest percentage of purchases with discounts applied?


```python
df['discount_flag']= df['discount_applied'].apply(lambda x: 1 if x=='Yes' else 0)
resulte=df.groupby('item_purchased').agg(total_count=('item_purchased','count'),discount=('discount_flag','sum'))
resulte['dis']=(resulte['discount']/resulte['total_count'])*100
resulte=resulte.reset_index()
```

```python
resulte[['item_purchased','dis']].sort_values('dis',ascending=False).head(5).round(2)
```

```python
g = df.groupby('item_purchased').agg({'item_purchased':'count','discount_flag':'sum'})
g['dis']= (g['discount_flag'])/g['item_purchased']*100
g.sort_values('dis',ascending=False).head(5).round(2)
```

## fast way

```python
g2 = df.groupby("item_purchased")[['discount_flag']].mean().sort_values(by='discount_flag' ,ascending =False ).round(4).reset_index().head()
g2['discount_flag'] = g2['discount_flag']*100
g2
```

## Segment customers into New, Returning, and Loyal based on their total  number of previous purchases, and show the count of each segment. 

```python
df['customer_segment']=df['previous_purchases'].apply(lambda x: 'new' if x<=1 else ('Returning' if x>=2 and x <=10 else 'Loyal'))
```

```python
df['customer_segment'].value_counts().reset_index()
```

### What are the top 3 most purchased products within each category? 


```python
top3 = df.groupby(['category','item_purchased'])['customer_id'].count().reset_index()
top3['rank']=top3.groupby(['category'])['customer_id'].rank(method='first', ascending=False)
```

```python
top3[(top3['rank'] <=3)].sort_values(['category','rank'])
```

### What are the top 3 most purchased products amounts within each category? 


```python
top3amount = df.groupby(['category','item_purchased'])['purchase_amount'].sum()
top3amount=top3amount.reset_index()
top3amount
```

```python
top3amount['rank'] = top3amount.groupby('category')['purchase_amount'].rank(method='first' , ascending=False)
```

```python
top3amount[(top3amount['rank'] <= 3)].sort_values(['category','rank'])
```
